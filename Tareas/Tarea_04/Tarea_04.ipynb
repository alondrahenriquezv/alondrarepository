```python
# Importar Matplot
import matplotlib.pyplot as plt

# Definir variables, valores y colores

fig, ax = plt.subplots()

days = ['Lunes', 'Martes', 'Miércoles','Jueves']
counts = [8, 7, 7, 6]
bar_labels = ['red', 'blue', 'orange', 'pink']
bar_colors = ['tab:red', 'tab:blue', 'tab:orange', 'tab:pink']

ax.bar(days, counts, label=bar_labels, color=bar_colors)

# Definir variable 'Y' + título del gráfico
ax.set_ylabel('Horas')
ax.set_title('Cantidad de horas de sueño')


plt.show()
```
